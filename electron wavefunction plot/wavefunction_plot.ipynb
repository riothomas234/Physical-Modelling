```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from numpy import pi
from mpl_toolkits.mplot3d import Axes3D
x= np.arange( -1.5, 1.5, 0.1)
t= np.arange( -0.5 , 2.5, 0.1)
xx, tt = np.meshgrid(x,t)

psi= 0.5*(np.sin(2*pi*(xx-tt))/(2*pi*(xx-tt))+np.sin(2*pi*(xx+tt))/(2*pi*(xx+tt)))



# Set font for labels in coming figures
font16 = {'family': 'serif',
          'color':  'black',
          'weight': 'normal',
          'size': 16,
          }

# Set font for labels in coming figures
font20 = {'family': 'serif',
          'color':  'black',
          'weight': 'normal',
          'size': 20,
          }


ax = plt.axes(projection='3d')
ax.plot_surface(xx,tt,psi, cmap=cm.plasma)
# Controlling appearance of ticks and labels
plt.xticks(np.arange(-2, 2, step=1),fontsize=16,fontfamily='serif') # style={'normal', 'italic', 'oblique'}
plt.yticks(np.arange(-1, 3, step=1),fontsize=16,fontfamily='serif')
plt.xlabel('x',fontdict=font16)
ax.set_zticks(np.arange(-0.2, 1.2, 0.2))
plt.ylabel('t',fontdict=font16)
ax.set_zlabel('$z$', rotation=180) # plt.zlabel('z')
plt.title('wavefunction',fontdict=font20)
ax.view_init(60, 90)
fig1 = plt.figure()

plt.show()




x= np.arange( -1.5, 1.5, 0.1)
t= np.arange( -0.5 , 2.5, 0.1)
xx, tt = np.meshgrid(x,t)

psi= 0.5*(np.sin(2*pi*(xx-tt))/(2*pi*(xx-tt))+np.sin(2*pi*(xx+tt))/(2*pi*(xx+tt)))




ax = plt.axes(projection='3d')
plt.xlabel('x',fontdict=font16)
plt.ylabel('t',fontdict=font16)
ax.set_zlabel('$z$', rotation=180)
plt.title('wavefunction',fontdict=font20)

ax.plot_wireframe(xx,tt,psi)
ax.view_init(60, 90)
fig = plt.figure()
plt.show()
```

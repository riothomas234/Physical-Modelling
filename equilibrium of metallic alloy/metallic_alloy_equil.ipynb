```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize




#critical temperature set.
t_c=1.


#setting up the range of temperature values over which the equilibrium q is plotted against.
num_t=200
initial_t=0.5
final_t=1.2

#setting up the range of q values over which free energy is plotted

num_q = 200
initial_q=-1
final_q=1
q = np.linspace(initial_q,final_q,num_q)


#function that calculates f(q)
def bw(q):
  f = -t_c*q**2+t*((1+q)*np.log(1+q)+(1-q)*np.log(1-q))
  return f

#plot f against q for 3 different temperatures
temps = [0.8,1.0,1.4]
for t in temps:
  t=t
  plt.plot(q,bw(q))



plt.ylabel('Free energy')
plt.xlabel('Q')
plt.legend(['T=0.8','T=1.0','T=1.4'])
plt.show()



q_0=0.9
q_e=[]

#finding the q for which f is minimised over a range of temperatures
temps = np.linspace(initial_t,final_t,num_t)
for t in temps:
  t=t
  res =minimize(bw,q_0,method='nelder-mead', options={'xatol': 1e-8})
  q_e.append(res.x)


#plot equilibrium_q against temperature.
plt.plot(q_e,temps, linewidth=3)
plt.ylabel('T')
plt.xlabel('Equilibrium Q')
plt.show()
#xatol sets the limit to how much x-values between iterations can change before we consider the x-value to have converged.
```

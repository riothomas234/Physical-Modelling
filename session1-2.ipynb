```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt


#create array of x values
x = np.linspace(-10, 10, 100)
#plots x against y for 3 different values of a
for i in range(1, 4):
  a = i
  y = x**3+a*x**2+3*x+4
  plt.plot(x,y)

#creates the graph
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Cubic curves')
plt.show()
```

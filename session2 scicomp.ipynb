```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('shares.dat', delimiter=',',skip_header=2)
year = data[:,0]
apple_price = data[:,1]
IBM_price = data[:,2]
microsoft_price=data[:,3]

plt.plot(year, apple_price, label='Apple')
plt.plot(year, IBM_price, label='IBM')
plt.plot(year, microsoft_price, label='Microsoft')
plt.xlim(2000, 2016)
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.legend()
plt.title('Stock prices against time')
plt.savefig('stockpricefigure.png', dpi=300)
plt.show()
```
